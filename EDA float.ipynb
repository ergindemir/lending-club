## Replace missing values with mean

```python
df_fpmean = df.mean()
```

```python
df_fpmean
```

```python
np.sum(df.isnull())
```

```python
for colname in df.columns:
    df[colname][df[colname].isnull()]=df_fpmean[colname]
```

```python
np.sum(df.isnull())
```
